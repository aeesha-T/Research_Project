```python
import pandas as pd
import numpy as np
import matplotlib as plt
import os
```

```python
table = pd.read_csv("../MDVR_acoustic_features_chunks.csv")
df = table
print("Total number of .wav files: ", len(df))
```

```python
df
```

```python
IDs = []
for i in range(len(df)):
    x = os.path.dirname(df["voiceID"][i]).split("\\")[3]
    IDs.append(x)
df['IDs'] = IDs
df
```

```python
print("Total number of .wav files: ", len(df))
print ("Size of HC and PD", df.groupby(['label']).size())

print("Average number of segments per sample is: ", df['IDs'].value_counts().describe()["mean"],
     " with std deviation of: ",df['IDs'].value_counts().describe()["std"])

df['IDs'].value_counts()
```
